## Limpieza de los datos

```python
#Eliminar columnas que no me interesan. Primero establezco tres condiciones sobre el nombre de las columnas que
#quiero eliminar. Las que empiecen por "Metro" y las que contengan "Urban" y "change"
eliminar=df_final.columns[df_final.columns.str.startswith("Metro")]
eliminar1=df_final.loc[:,df_final.columns.str.contains("Urban")]
eliminar2=df_final.loc[:,df_final.columns.str.contains("change")]

df_final=df_final\
.drop(columns=eliminar)\
.drop(columns=eliminar1)\
.drop(columns=eliminar2)\

df_final

```

```python
#estrechamos la tabla
df_final=df_final.melt(id_vars=["FIPS", "Stabr","County_Name","geoid"], var_name= "prodyear", value_name= "valor")
df_final
```

```python
#separamos los valores de la columna "variable" en dos columnas "product" y en "year". Para ello separo 
#letras de numeros utilizando expresiones regulares. Primero guardo en las variables "numeros" y "letras" la 
#expresión regular, y después las utilizo en las values de la columna "prodyear".
#Importante: al crear la columna "year", convertirla en integer.
#Por último elimino la columna "prodyear"


numeros= r'(\d+)'
letras= r"\b[^\d\W]+"

df_final=(df_final.assign(
    year= lambda dataset: dataset["prodyear"].map(lambda value: int(re.search(numeros, value).group())),
    product= lambda dataset: dataset["prodyear"].map(lambda value: re.search(letras, value).group())
       
)
 .drop(columns="prodyear")
 
)

df_final
```

```python
#Cambiamos el nombre de las columnas a minúsculas
df_final=df_final.rename(columns= lambda column: column.lower())
df_final
```

```python
#Eliminamos las filas que tienen en la columna "valor" el 0
df_final= df_final.query("valor != 0")
df_final
```

```python
#Ensanchamos la tabla, de tal manera que los valores de la columna "product" se conviertan en columnas "gas"
#y "oil"
df_final=df_final.pivot(index=["fips", "stabr", "county_name", "geoid", "year"], 
                        columns= "product", 
                        values= "valor"
                       )\
.reset_index()\
.rename_axis(columns=None)

df_final
```
