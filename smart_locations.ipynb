## Limpieza de datos

```python
#nos quedamos solo con las columnas que nos interesan
df_uno= df_uno.filter(['id', 'positions'])
df_uno.head()
```

```python
#convierto las strings de positions en JSON para luego poder acceder a los valores json.loads(loquesea)
import json

df_uno=df_uno.assign(
    positions = lambda dataset: dataset["positions"].map(lambda value: json.loads(value))
)
df_uno.head()
```

```python
#utilizo el método explode() para que cada array de value ocupe una fila
df_uno=df_uno.explode('positions')
df_uno
```

```python
df_uno=df_uno.reset_index(drop=True)
df_uno
```

```python
#comprobamos que está todo correcto
df_uno.iat[1,1]
```

```python
#guardo los datos de cada fila de postions en un dataframe nuevo
df_dos=pd.json_normalize(df_uno['positions'])
df_dos
```

```python
df_final=pd.concat([df_uno, df_dos], axis=1)
df_final
```

```python
df_final=df_final\
.drop(columns= ['positions'])\
.rename(columns= {'id': 'Id'})

df_final.head()

```
